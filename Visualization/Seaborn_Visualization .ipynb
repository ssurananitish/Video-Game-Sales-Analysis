# Preparation and Data Cleaning

Importing the ploting packages required

```python
import pandas as pd;
import numpy as np;
from matplotlib import pyplot as plt
import seaborn as sns
```

Reading and exploring data

```python
data = pd.read_csv('vgsales.csv')
data.info()

#Categorical data -> Platform,Genre,Publisher, Year
#Ratio data -> NA_Sales,EU_Sales, JP_Sales, Other_Sales, Global_Sales
#No interval data 
```

Dropping all NA entries and converting Year to Integer format.

```python
data.dropna(inplace=True);
data['Year'] = data['Year'].astype(int)
data.info()
```

```python
data.Platform.unique()
```

# Exploratory Data Analysis (EDA)

I have attached an cumulative list having all the inferences taken from the EDA at the end.

What are the global sales with respect to different type of platforms?

```python
platform_unique=list(data.Platform.unique())
platform_sales=[]
for i in platform_unique:
    x = data[data['Platform']==i]
    #print(x)
    platform_total = sum(x.Global_Sales)
    platform_sales.append(platform_total)
      
data2 = pd.DataFrame({'platform_unique': platform_unique,'platform_sales':platform_sales})
new_index = (data2['platform_sales'].sort_values(ascending=False)).index.values
sorted_data = data2.reindex(new_index)

# visualization
plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data['platform_unique'], y=sorted_data['platform_sales'])
plt.xlabel('Platform')
plt.ylabel('Global Sales')
plt.title('Global Sales Given Platform')
```

How are ths global sales spread out over the years?

```python
for i in genre_unique:
    plt.figure(figsize=(15,10))
    x=data[data['Genre']==i]
    
    sns.distplot(x['Global_Sales'], hist = False, kde = True,kde_kws = {'linewidth': 2},label=i)
    
#plt.legend(labels)
plt.title('Density Plot with Multiple Platforms')
plt.xlabel('Global Sales')
plt.ylabel('Density')
```

Show the comparison between the sales of different regions along the years.

```python
year_JP_sales=[]
year_NA_sales=[]
year_EU_sales=[]
year_Other_sales=[]
year_unique=list(data.Year.unique())
year_unique.sort()

for i in year_unique:
    x=data[data['Year']==i]
    year_total = sum(x.JP_Sales)
    year_JP_sales.append(year_total)
    year_total = sum(x.NA_Sales)
    year_NA_sales.append(year_total)
    year_total = sum(x.EU_Sales)
    year_EU_sales.append(year_total)
    year_total = sum(x.Other_Sales)
    year_Other_sales.append(year_total)
    
plt.figure(figsize=(15,10))
plt.plot(year_unique,year_NA_sales,label="NA_Sales")
plt.plot(year_unique,year_EU_sales,label="EU_Sales")
plt.plot(year_unique,year_JP_sales,label="JP_Sales")
plt.plot(year_unique,year_Other_sales,label="Other_Sales")
plt.grid()
plt.legend()
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales per Year')
plt.show()
```
